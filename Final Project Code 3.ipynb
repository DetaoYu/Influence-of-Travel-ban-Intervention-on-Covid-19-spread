```python
import pandas as pd
```

```python
flight = pd.read_csv('flightlist_20200101_20200131.csv')
flight.drop(columns=['callsign', 'number', 'icao24', 'registration', 'typecode'], inplace=True)
flight
```

```python
airports = pd.read_csv('airports.dat', names=['Airport ID', 'Name', 'City', 'Country', 'IATA', 'ICAO', 'Latitude', 'Longitude', 'Altitude', 'Timezone', 'DST', 'Tz database time zone', 'Type', 'Source'])
airports.drop(columns = ['Airport ID', 'Name', 'City', 'Country', 'IATA', 'Altitude', 'Timezone', 'DST', 'Tz database time zone', 'Type', 'Source'], inplace=True)
# airports
```

```python
airports['Coordinates'] = pd.Series(zip(airports['Latitude'], airports['Longitude']))
```

```python
airports.head()
```

```python
airports.to_csv('airports.csv')
```

```python
data = pd.merge(left=flight, right=airports, left_on='origin', right_on='ICAO')
```

```python
data
```

```python
data.rename(columns={'Latitude': 'Latitude1', 'Longitude': 'Longitude1'}, inplace=True)
```

```python
datas = pd.merge(left=data, right=airports, left_on='destination', right_on='ICAO')
```

```python
datas.to_csv('datas.csv')
```
